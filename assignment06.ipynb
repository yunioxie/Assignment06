## 01-1)

풀이과정 : 리프노드는 7, 17, 70이다.  
답 : 3개

## 01-2)

풀이과정 : 높이가 4일 때 리프노드는 최대 2^(4-1)=8이다. 현재 17이 있으므로 7개를 추가할 수 있다.  
답 : 7개

## 01-3)

풀이과정 : 리프노드의 개수는 2^(4-1)=8이고 전체 노드는 2^4-1=15로 내부노드의 개수는 15-8=7이다. 리프노드와 내부노드의 개수의 차이는 1개다.  
답 : 1개

## 01-4)

풀이과정 : 높이가 h인 모든 포화 이진 트리에서 전체 노드의 개수는 2^h-1이다. 리프노드의 개수는 2^(h-1)이고 내부노드는 (2^h-1)-(2^(h-1))=2^(h-1)-1이다. 리프노드와 내부노드의 개수의 차이는 2^(h-1)-(2^(h-1)-1)=1이다. 따라서 모든 포화 이진 트리에서 성립한다.  
답 : O

## 02-1)

풀이과정 : 총 노드 수가 n이면 높이는 log₂(n+1)이다. 높이가 log₂(n+1)-1인 포화 이진 트리의 총 노드 수는 2^(log₂(n+1)-1)-1 = (n-1)/2이다. 따라서 리프 노드 수는 n - (n-1)/2 = (n+1)/2이다.  
답 : (n+1)/2

## 02-2)

풀이과정 : 리프 노드 수는 (n+1)/2이고 내부 노드 수는 n - (n+1)/2 = (n-1)/2이다. 리프 노드 수가 내부 노드 수보다 1개 더 많다.  
답 : 리프 노드의 수가 내부 노드의 수보다 1개 더 많다.

## 03-4)

풀이 :  
가정 : k < h인 모든 k에 대해 성립한다. (높이 = h)  
h = 1일 때 트리는 1개의 노드를 가지고 노드 1개를 방문하는 것은 정렬 순서로 방문하는 것과 같다.  
h >= 2일 때 중위 순회는 왼쪽 서브 트리, 오른쪽 서브 트리를 각각 정렬된 순서로 방문한다. 왼쪽 서브트리의 모든 키는 루트의 키보다 작고 오른쪽 서브트리의 모든 키는 루트의 키보다 크다. 따라서 중위 순회는 정렬 순서로 방문한다.
